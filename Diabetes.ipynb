**Giải thích cơ bản về bệnh tiểu đường:**

Bệnh tiểu đường - đái tháo đường hay đúng hơn là rối loạn đường huyết do hormone Insulin.

Insulin được coi như chìa khoá mở cửa tế bào để đưa đường-glucose trong máu vào và sử dụng để tạo ra năng lượng sống. Khi đó đường không vào trong tế bào nên tế bào bị thiếu năng lượng sau đó tế bào tự ăn chính nó.

Triệu chứng thường thấy ở người bệnh: Gầy, sụt cân nhanh, ăn nhiều vẫn đói, đi tiểu nhiều, vết thương lâu lành.

Căn bệnh này nếu không được phát hiện sớm sẽ dẫn đến nhiều biến chứng nghiêm trọng như tổn thương thận(tiểu ra đường), tim, các cơ quan, tổn thương thần kinh,...

Chuẩn đoán: đo đường huyết khi đói hoặc thực hiện các xét nghiệm máu chuyên sâu.

Tiểu đường được phân ra làm 2 loại là tiểu đường Type 1 và Type 2. Trong đó:

- Tiểu đường Type 1: Nguyên nhân do bẩm sinh không có insulin hoặc tuyến tụy không tiết được insulin.

  Biện pháp: Sử dụng insulin ngoại sinh suốt đời

- Tiểu đường Type 2: Nguyên nhân do đề kháng insulin(do nạp nhiều đường trong thời gian dài), tăng đường huyết liên tục, béo phì làm cơ thể phải tăng lượng insulin cao hơn bình thường để đưa đường vào tế bào

  Biện pháp: Sử dụng thuốc tăng tiết insulin và giảm đề kháng insulin

**Chia dữ liệu thành data train và data test với tỉ lệ train 80% và test 20%**

```python
X = df.drop(columns='Diabetes',axis = 1)
Y = df['Diabetes']
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size=0.2,random_state=42)
```

```python
print("Data gốc",X.shape)
print("Data train",X_train.shape)
print("Data test",X_test.shape)
```

**Model traning**

Decision Tree Classifier

```python
DTC = DecisionTreeClassifier()
DTC.fit(X_train, Y_train)
DTC_predict = DTC.predict(X_test)
print("Độ chính xác: ",100* accuracy_score(Y_test,DTC_predict ))
```

K-Neighbors Classifier

```python
KNN = neighbors.KNeighborsClassifier(n_neighbors=10, weights='distance')
KNN.fit(X_train, Y_train)
KNN_predict = KNN.predict(X_test)
print("Độ chính xác",100*accuracy_score(Y_test,KNN_predict))
```

**Đánh giá Model**

Decision Tree Classifier

```python
print('Accuracy Score:', accuracy_score(Y_test, DTC_predict))
print('Precision Score:', precision_score(Y_test, DTC_predict))
print('Recall Score', recall_score(Y_test, DTC_predict))
```

K-Neighbors Classifier

```python
print('Accuracy Score:', accuracy_score(Y_test, KNN_predict))
print('Precision Score:', precision_score(Y_test, KNN_predict))
print('Recall Score', recall_score(Y_test, KNN_predict))
```

So sánh 2 Model

```python
plt.figure(figsize=(10, 5))
plt.bar(['Decision Tree Classifier','K-Neighbors Classifier'], [accuracy_score(Y_test, DTC_predict), accuracy_score(Y_test, KNN_predict)], color=['skyblue', 'orange'])
plt.title('So sánh Accuracy của hai mô hình')
plt.ylabel('Accuracy')
plt.ylim(0, 1)
plt.show()
```
